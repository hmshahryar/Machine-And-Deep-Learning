With Error and train test split of l-r


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split


data_set = pd.read_csv("Polynomial-Regression-data.csv")


data_set["Salary"] = data_set["Salary"].replace(",", "", regex=True).astype(float)


X = data_set.iloc[:, 1:-1].values
y = data_set.iloc[:, -1].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)


regressor = DecisionTreeRegressor(max_depth=4, random_state=0)
regressor.fit(X_train, y_train)


predicted_salary = regressor.predict([[6.5]])
print(f"Predicted Salary for Position Level 6.5: ${predicted_salary[0]:,.2f}")


y_pred = regressor.predict(X_test)


mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
print(f"R² Score: {r2:.4f} (Closer to 1 is better)")


X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))


plt.figure(figsize=(10, 5))
plt.scatter(X, y, color="red", label="Actual Data")
plt.plot(X_grid, regressor.predict(X_grid), color="blue", linewidth=2, label="Decision Tree Prediction")
plt.title("Decision Tree Regression (Advanced)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.legend()
plt.grid()
plt.show()
```

With plotly


```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split


data_set = pd.read_csv("Polynomial-Regression-data.csv")

data_set["Salary"] = data_set["Salary"].replace(",", "", regex=True).astype(float)


X = data_set.iloc[:, 1:-1].values
y = data_set.iloc[:, -1].values


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)


regressor = DecisionTreeRegressor(max_depth=4, random_state=0)
regressor.fit(X_train, y_train)


predicted_salary = regressor.predict([[6.5]])
print(f"Predicted Salary for Position Level 6.5: ${predicted_salary[0]:,.2f}")

y_pred = regressor.predict(X_test)


mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
print(f"R² Score: {r2:.4f} (Closer to 1 is better)")

X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))


fig = go.Figure()


fig.add_trace(go.Scatter(x=X.flatten(), y=y, mode="markers", marker=dict(color="red", size=8), name="Actual Data"))


fig.add_trace(go.Scatter(
    x=X_grid.flatten(),
    y=regressor.predict(X_grid),
    mode="lines",
    line=dict(color="blue", width=2),
    name="Decision Tree Prediction"
))


fig.update_layout(
    title="Interactive Decision Tree Regression",
    xaxis_title="Position Level",
    yaxis_title="Salary",
    template="plotly_white"
)


fig.show()
```
