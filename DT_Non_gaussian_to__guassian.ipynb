```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(0)
df = np.random.exponential(size=1000 , scale=2)
df = pd.DataFrame(df, columns=['values'])
df
```

```python
sns.histplot(df['values'], kde=True)
plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import QuantileTransformer

qt_normal = QuantileTransformer(output_distribution='normal')
pt_box_cox = PowerTransformer(method='box-cox') # data must be positive 
pt_yeo_johnson = PowerTransformer(method='yeo-johnson')
```

```python
df["box_cox"] = pt_box_cox.fit_transform(df[["values"]]+1)
df["yeo_johnson"] = pt_yeo_johnson.fit_transform(df[["values"]])
df["quantile_normal"] = qt_normal.fit_transform(df[["values"]])
```

```python
sns.histplot(df['box_cox'], kde=True)
plt.show()

sns.histplot(df['yeo_johnson'], kde=True)
plt.show()

sns.histplot(df['quantile_normal'], kde=True)
plt.show()
```
